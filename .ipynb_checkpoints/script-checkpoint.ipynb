```python
import pandas as pd
```

```python
df = pd.read_excel('Copy of 2018 ASP RankingF (00000002).xlsx', sheet_name='Labor', usecols='J,AC')
```

```python
size = df.groupby(df.columns.tolist()).size()
temp = size.to_frame(name = 'size').reset_index()
temp
```

```python
temp = temp[temp['size']>1]
res = temp.groupby('Customer Name').size()
df_output = res.to_frame(name = '>1').reset_index()
df_output
```

```python
count = df.groupby('Customer Name')['Serial_Number'].count()
df_count = count.to_frame(name = 'count').reset_index()
df_count
```

```python
df_output.to_csv('Result.csv', index=False)
```
